```python
hdu  = fits.open('shape_mode_meanify.fits')
xy = hdu[1].data['COORDS0'][0]
size = hdu[1].data['PARAMS0'][0][:,0]
g1 = hdu[1].data['PARAMS0'][0][:,1]
g2 = hdu[1].data['PARAMS0'][0][:,2]
```

```python
x = xy[:,0]
y = xy[:,1]
print(x.shape)
```

```python
f,ax = plt.subplots(1,1,figsize=(15,12))

xmin = x.min()
xmax = x.max()
ymin = y.min()
ymax = y.max()

hb = ax.hexbin(x, y, C=size, gridsize=150, cmap=plt.cm.RdBu,vmax=0.04, vmin=-0.04)
ax.axis([xmin, xmax, ymin, ymax])
ax.set_title("atmo_size")
cb = f.colorbar(hb, ax=ax)
```
